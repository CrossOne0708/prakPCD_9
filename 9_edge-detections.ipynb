```python
# Mengimport library yg digunakan
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import data
```

## **1) Proses Sobel Filter**

```python
# Load gambar dari local drive
image = cv2.imread("images/raze.jpg")
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Sobel
image_sobelx = cv2.Sobel(grayimage, cv2.CV_8U, 1, 0, ksize=5)
image_sobely = cv2.Sobel(grayimage, cv2.CV_8U, 0, 1, ksize=5)
image_sobel = image_sobelx + image_sobely

# Plotting Output Image
fig, axes = plt.subplots(4, 2, figsize=(15, 15))
ax = axes.ravel()

# Plot Citra Input
ax[0].imshow(grayimage, cmap='gray')
ax[0].set_title("Citra Input")
ax[1].hist(grayimage.ravel(), bins = 256)
ax[1].set_title("Historam Citra Input")

# Plot Output Sobel X
ax[2].imshow(image_sobelx, cmap='gray')
ax[2].set_title("Citra Output Sobel X")
ax[3].hist(image_sobelx.ravel(), bins = 256)
ax[3].set_title("Historam Citra Output Sobel X")

# Plot Output Sobel Y
ax[4].imshow(image_sobely, cmap='gray')
ax[4].set_title("Citra Output Sobel Y")
ax[5].hist(image_sobely.ravel(), bins = 256)
ax[5].set_title("Historam Citra Output Sobel Y")

# Plot Output Sobel
ax[6].imshow(image_sobel, cmap='gray')
ax[6].set_title("Citra Output Sobel X & Y")
ax[7].hist(image_sobel.ravel(), bins = 256)
ax[7].set_title("Historam Citra Output Sobel X & Y")

fig.tight_layout()
plt.show()
```

## **2) Proses Prewitt Filter**

```python
# Load gambar dari local drive
image = cv2.imread("images/raze.jpg")
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

image_prewittx = cv2.filter2D(grayimage, -1, kernelx)
image_prewitty = cv2.filter2D(grayimage, -1, kernely)
image_prewitt = image_prewittx + image_prewitty

# Plotting Output Image
fig, axes = plt.subplots(4, 2, figsize=(15, 15))
ax = axes.ravel()

# Plot Citra Input
ax[0].imshow(grayimage, cmap='gray')
ax[0].set_title("Citra Input")
ax[1].hist(grayimage.ravel(), bins = 256)
ax[1].set_title("Historam Citra Input")

# Plot Output Prewitt X
ax[2].imshow(image_prewittx, cmap='gray')
ax[2].set_title("Citra Output Prewitt X")
ax[3].hist(image_prewittx.ravel(), bins = 256)
ax[3].set_title("Historam Citra Output Prewitt X")

# Plot Output Prewitt Y
ax[4].imshow(image_prewitty, cmap='gray')
ax[4].set_title("Citra Output Prewitt Y")
ax[5].hist(image_prewitty.ravel(), bins = 256)
ax[5].set_title("Historam Citra Output Prewitt Y")

# Plot Output Prewitt
ax[6].imshow(image_prewitt, cmap='gray')
ax[6].set_title("Citra Output Prewitt X & Y")
ax[7].hist(image_prewitt.ravel(), bins = 256)
ax[7].set_title("Historam Citra Output Prewitt X & Y")

fig.tight_layout()
plt.show()
```

## **4) Proses Filter Turunan Kedua (Laplacian & Gaussian)**

```python
# Load gambar dari local drive
image = cv2.imread("images/raze.jpg")
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplikasikan gaussian
blur = cv2.GaussianBlur(grayimage, (3,3), 0)

# Konvolusi dengan kernel
laplacian = cv2.Laplacian(grayimage, cv2.CV_64F)

# Plotting Output
plt.subplot(1,2,1), plt.imshow(grayimage, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
```

```python
# Load gambar dari local drive
image = cv2.imread("images/raze.jpg")
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplikasikan gaussian
blur = cv2.GaussianBlur(grayimage, (3,3), 0)

# Konvolusi dengan kernel
laplacian = cv2.Laplacian(blur, cv2.CV_64F)
laplacian1 = laplacian/laplacian.max()

plt.imshow(laplacian1, cmap='gray'), plt.axis('off')
plt.title('Laplacian-Gaussian')
plt.show()
```
